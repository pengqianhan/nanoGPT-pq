```python
### 学习率打印
import math
import matplotlib.pyplot as plt

def get_lr(it):
    # Parameters
    warmup_iters = 100
    lr_decay_iters = 5000
    learning_rate = 1e-4
    min_lr = 1e-5

    # 1) linear warmup for warmup_iters steps
    if it < warmup_iters:
        return learning_rate * it / warmup_iters
    # 2) if it > lr_decay_iters, return min learning rate
    if it > lr_decay_iters:
        return min_lr
    # 3) in between, use cosine decay down to min learning rate
    decay_ratio = (it - warmup_iters) / (lr_decay_iters - warmup_iters)
    assert 0 <= decay_ratio <= 1
    coeff = 0.5 * (1.0 + math.cos(math.pi * decay_ratio)) # coeff ranges 0..1
    return min_lr + coeff * (learning_rate - min_lr)

# Generate data points
iterations = range(0, 6000)
learning_rates = [get_lr(it) for it in iterations]

# Plot the curve
plt.figure(figsize=(12, 6))
plt.plot(iterations, learning_rates)
plt.title('Learning Rate Schedule')
plt.xlabel('Iterations')
plt.ylabel('Learning Rate')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()

# Print some key points
print(f"Initial LR: {get_lr(0)}")
print(f"LR at warmup_iters: {get_lr(100)}")
print(f"LR at middle of cosine decay: {get_lr(2550)}")
print(f"Final LR: {get_lr(6000)}")
```
